```python
with open('/Users/chughes/Downloads/Web_Scraping_and_Text_Analysis_in_Python/USB/Data/Clean_BBC_Business_Articles.json', 'r') as f:
    data = json.load(f)
```

```python
data[0]
```

```python
import gensim
import gensim.corpora as corpora
```

```python
bow_1 = corpora.Dictionary(data)
```

```python
bow_1[0]
```

```python
bow_1.id2token
```

```python
bow_2 = []
for document in data:
    document_out = bow_1.doc2bow(document)
    bow_2.append(document_out)
```

```python
lda_model = gensim.models.ldamodel.LdaModel(
    corpus=bow_2,
    id2word=bow_1,
    num_topics=20, 
    passes=5,
)
```

### Evaluating LDA

```python
lda_model.print_topic(0)
```

```python
topics = lda_model.print_topics(num_words=6)
```

```python
for topic in topics:
    print(topic)
```

```python
unseen_data = ['crude', 'oil', 'price', 'drops', 'dollar', 'interest', 'rates', 'bank']
```

```python
unseen_bow = bow_1.doc2bow(unseen_data)
print(lda_model.get_document_topics(unseen_bow))
```

```python
lda_model.get_document_topics(unseen_bow,minimum_probability=0.0)
```
